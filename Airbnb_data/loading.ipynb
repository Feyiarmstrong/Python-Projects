Import Libraries

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import requests
from io import BytesIO
import gzip
import os
```

Load Data

```python
file_path = "data/london_listing_raw.csv" 
url = "https://data.insideairbnb.com/united-kingdom/england/london/2025-09-14/data/listings.csv.gz"

# Download and read directly into DataFrame
response = requests.get(url)
with gzip.open(BytesIO(response.content), 'rt') as f:
    df_raw = pd.read_csv(f)

# Save to local CSV (overwrite if exists)
df_raw.to_csv(file_path, index=False)
print("Saved file:", file_path)
```
