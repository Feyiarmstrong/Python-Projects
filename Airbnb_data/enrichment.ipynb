Import libraries

```python
import pandas as pd
from src.enricher import enrich_airbnb_data
```

Load Data

```python
df_clean = pd.read_csv("data/london_listing_clean.csv")
df_enriched = enrich_airbnb_data(df_clean)
```

```python
df_enriched.head()
```

Create new columns

```python
# Create price_per_booking column
df_enriched['price_per_booking'] = df_enriched['price'] * df_enriched['minimum_nights']

# Preview the change
df_enriched[['price', 'minimum_nights', 'price_per_booking']].head()
```

```python
# Create availability category
df_enriched['availability_category'] = pd.cut(
    df_enriched['availability_365'],
    bins=[-1, 99, 300, 365],
    labels=['Rare', 'Part-time', 'Full-time']
)
```
