Import Libraries

```python
from src.analyzer import load_enriched_data
```

```python
#Load the enriched dataset
df_enriched = load_enriched_data()
```

```python
df_enriched.head()
```

```python
#Top 10 most expensive neighborhoods by average price
top_neighborhoods = (
    df_enriched.groupby('neighbourhood_cleansed')['price']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

top_neighborhoods
```

```python
# Average availability and price by room type
avg_by_room_type = (
    df_enriched.groupby('room_type')[['availability_365', 'price']]
    .mean()
    .reset_index()
)

avg_by_room_type
```

```python
# Host with the most listings
top_host = (
    df_enriched['host_name']
    .value_counts()
    .head(1)
)

top_host
```

```python
# Average price by borough/district
avg_price_borough = (
    df_enriched.groupby('neighbourhood_cleansed')['price']
    .mean()
    .sort_values(ascending=False)
)

avg_price_borough.head(5)
```

```python
# Listings with zero reviews
never_reviewed = df_enriched[df_enriched['number_of_reviews'] == 0]
num_never_reviewed = never_reviewed.shape[0]

num_never_reviewed
```

### Key Insights from Analysis

- The most expensive neighborhoods are City of London, Lambeth, Kensington and Chelsea.

- Entire home/apartments have higher average prices and availability than private rooms.

- A few hosts control a significant portion of listings, dominating the market.

- Many listings have never been reviewed, indicating inactive or new listings.

- Most listings are “Full-time” or “Part-time” in availability, with few “Rare” listings.
